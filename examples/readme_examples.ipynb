```julia
using LibGEOS
```

```julia
p1 = parseWKT("POLYGON((0 0,1 0,1 1,0 0))")
p2 = parseWKT("POLYGON((0 0,1 0,1 1,0 1,0 0))")
p3 = parseWKT("POLYGON((2 0,3 0,3 1,2 1,2 0))")
```

# Compose and Color

```julia
using Compose, Gadfly, Color
```

```julia
colors = Color.colormap("Greens", 3)
```

```julia
plot(x=[-0.5,3.5],y=[-0.5,1.5], Geom.point,
     Guide.annotation(compose(context(),GeoInterface.composeform(p1),linewidth(0.05mm),stroke(color("black")),fill(colors[1]))),
     Guide.annotation(compose(context(),GeoInterface.composeform(p3),linewidth(0.05mm),stroke(color("black")),fill(colors[2]))),
     Guide.annotation(compose(context(),GeoInterface.composeform(p2),linewidth(0.05mm),stroke(color("black")),fill(colors[3]))))
```

```julia
plot(x=[-0.5,3.5],y=[-0.5,1.5], Geom.point,
     Guide.annotation(compose(context(),GeoInterface.composeform(buffer(p1,0.5)),linewidth(0.05mm),stroke(color("black")),fill(colors[1]))),
     Guide.annotation(compose(context(),GeoInterface.composeform(buffer(p3,0.5)),linewidth(0.05mm),stroke(color("black")),fill(colors[2]))),
     Guide.annotation(compose(context(),GeoInterface.composeform(buffer(p2,0.5)),linewidth(0.05mm),stroke(color("black")),fill(colors[3]))))
```

# Interact

```julia
using Interact

@manipulate for width=-0.5:0.1:0.5, quadsegs=1:10
    plot(x=[-0.5,3.5],y=[-0.5,1.5], Geom.point,
         Guide.annotation(compose(context(),GeoInterface.composeform(buffer(p1,width,quadsegs)),linewidth(0.05mm),stroke(color("black")),fill(colors[1]))),
         Guide.annotation(compose(context(),GeoInterface.composeform(buffer(p3,width,quadsegs)),linewidth(0.05mm),stroke(color("black")),fill(colors[2]))),
         Guide.annotation(compose(context(),GeoInterface.composeform(buffer(p2,width,quadsegs)),linewidth(0.05mm),stroke(color("black")),fill(colors[3]))))
end
```

# Operations
## union

```julia
polygon = union(buffer(p1,0.5),buffer(p3,0.5))
```

```julia
plot(x=[-0.5,3.5],y=[-0.5,1.5], Geom.point,
Guide.annotation(compose(context(),GeoInterface.composeform(polygon),linewidth(0.05mm),stroke(color("black")),fill(colors[1]))))
```
